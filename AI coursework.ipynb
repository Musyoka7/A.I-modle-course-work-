```python
import pandas as pd
import numpy as np

# Load the dataset with predicted traversal costs
grid_data = pd.read_csv('Estimated_grid.csv')

# Assume the grid is square (adjust if necessary)
num_cells = len(grid_data)
grid_size = int(np.sqrt(num_cells))  # e.g., 100 cells -> 10x10 grid

# Reshape predicted traversal costs into a 2D grid
cost_grid = grid_data['predicted_traversal_cost'].values.reshape((grid_size, grid_size))

# Print to verify
print("2D Grid of Traversal Costs:")
print(cost_grid)
```

```python
# Create the graph as a dictionary
graph = {}

# Iterate through the 2D grid
for i in range(grid_size):
    for j in range(grid_size):
        # Current node
        node = (i, j)
        graph[node] = []

        # Define possible moves (up, down, left, right)
        moves = [
            (i-1, j),  # Up
            (i+1, j),  # Down
            (i, j-1),  # Left
            (i, j+1)   # Right
        ]

        # Add valid neighbors to the graph
        for move in moves:
            x, y = move
            if 0 <= x < grid_size and 0 <= y < grid_size:  # Check bounds
                # Add edge to the neighbor with its traversal cost as the weight
                weight = cost_grid[x, y]
                graph[node].append(((x, y), weight))

# Print a sample node's neighbors to verify
print("Neighbors of (0, 0):", graph[(0, 0)])
```

```python
import heapq  # For priority queue

def dijkstra(graph, start, goal):
    # Priority queue: (cost, node)
    pq = [(0, start)]  # Start with the source node, cost is 0
    distances = {node: float('inf') for node in graph}  # Initialize distances as infinite
    distances[start] = 0  # Distance to the start node is 0
    previous_nodes = {node: None for node in graph}  # Track the path

    while pq:
        # Pop the node with the smallest cost
        current_cost, current_node = heapq.heappop(pq)

        # If we've reached the goal, stop searching
        if current_node == goal:
            break

        # Explore neighbors
        for neighbor, weight in graph[current_node]:
            new_cost = current_cost + weight
            if new_cost < distances[neighbor]:  # Found a shorter path
                distances[neighbor] = new_cost
                previous_nodes[neighbor] = current_node
                heapq.heappush(pq, (new_cost, neighbor))

    # Reconstruct the path
    path = []
    node = goal
    while node is not None:
        path.append(node)
        node = previous_nodes[node]
    path.reverse()  # Reverse the path to start from the source

    return distances[goal], path  # Return the cost and the path
```

```python
# Define the start and goal nodes
start = (0, 0)
goal = (grid_size-1, grid_size-1)

# Run Dijkstra's algorithm
cost, path = dijkstra(graph, start, goal)

# Print the results
print("Least-cost path:", path)
print("Total cost:", cost)
```

```python
import numpy as np
import random

# Define grid size and obstacles
grid_size = 5
obstacles = [(1, 1), (3, 3)]  # Adjusted for 0-based indexing
goal = (4, 4)  # Adjusted for 0-based indexing
start = (0, 0)

# Actions and rewards
actions = {0: 'up', 1: 'down', 2: 'left', 3: 'right'}
action_mapping = {'up': (-1, 0), 'down': (1, 0), 'left': (0, -1), 'right': (0, 1)}
rewards = np.full((grid_size, grid_size), -1)
for obstacle in obstacles:
    rewards[obstacle] = -10
rewards[goal] = 100
```

```python
# Initialize Q-table
q_table = np.zeros((grid_size, grid_size, len(actions)))

# Hyperparameters
alpha = 0.1  # Learning rate
gamma = 0.9  # Discount factor
epsilon = 0.1  # Exploration rate
episodes = 1000
max_steps = 100
```

```python
# Training loop
for episode in range(episodes):
    state = start
    total_reward = 0

    for _ in range(max_steps):
        # Choose action using epsilon-greedy policy
        if random.uniform(0, 1) < epsilon:
            action = random.choice(list(actions.keys()))  # Explore
        else:
            action = np.argmax(q_table[state[0], state[1]])  # Exploit

        # Determine next state
        move = action_mapping[actions[action]]
        next_state = (state[0] + move[0], state[1] + move[1])

        # Check boundaries and obstacles
        if next_state[0] < 0 or next_state[0] >= grid_size or next_state[1] < 0 or next_state[1] >= grid_size:
            next_state = state  # Stay in the same position
        if next_state in obstacles:
            reward = -10
        elif next_state == goal:
            reward = 100
        else:
            reward = 1

        # Update Q-value
        old_value = q_table[state[0], state[1], action]
        next_max = np.max(q_table[next_state[0], next_state[1]])
        q_table[state[0], state[1], action] = old_value + alpha * (reward + gamma * next_max - old_value)

        # Update state and reward
        state = next_state
        total_reward += reward

        if state == goal:
            break
```

```python
# Test the trained agent
def test_agent(episodes=100):
    total_rewards = []
    for _ in range(episodes):
        state = start
        total_reward = 0
        for _ in range(max_steps):
            action = np.argmax(q_table[state[0], state[1]])  # Always exploit
            move = action_mapping[actions[action]]
            next_state = (state[0] + move[0], state[1] + move[1])

            if next_state[0] < 0 or next_state[0] >= grid_size or next_state[1] < 0 or next_state[1] >= grid_size:
                next_state = state
            if next_state in obstacles:
                reward = -10
            elif next_state == goal:
                reward = 100
            else:
                reward = 1

            state = next_state
            total_reward += reward

            if state == goal:
                break
        total_rewards.append(total_reward)
    return np.mean(total_rewards)

average_reward = test_agent()
print(f"Average reward after training: {average_reward}")
```

```python
# Visualize the best action for each state
policy = np.zeros((grid_size, grid_size), dtype=str)
for i in range(grid_size):
    for j in range(grid_size):
        best_action = np.argmax(q_table[i, j])
        policy[i, j] = actions[best_action][0]  # Use first letter for simplicity
print("Learned Policy:")
print(policy)
```
